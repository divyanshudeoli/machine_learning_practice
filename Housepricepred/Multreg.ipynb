```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x=np.arange(1,501)
```

```python
x1=x*5-7
```

```python
x2=x*-2+119
```

```python
x3=(x**1.002)*7-(3*x)+15
```

```python
y=x1+x2+x3+np.random.normal(0,1000,500)
```

```python
plt.plot(x,y)
```

```python
xdf=pd.DataFrame(data=[x1,x2,x3]).T
```

```python
xdf.columns=['x1','x2','x3']
```

```python
xtr,xtest,ytr,ytest=train_test_split(xdf,y)
```

```python
y=y.reshape(-1,1)
```

```python
le=LinearRegression()
```

```python
le.fit(xtr,ytr)
```

```python
pred=le.predict(xtest)
```

```python
pred
```

```python
r2_score(pred,ytest)
```
