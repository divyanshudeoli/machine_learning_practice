```python
import pandas as pd
```

```python
import numpy as np
```

```python
train=pd.read_csv("train.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
train=train.apply(le.fit_transform)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
x=train.drop(['SalePrice'],axis=1)
```

```python
x
```

```python
y=train['SalePrice']
```

```python
bestfeature=SelectKBest(score_func=chi2)
fit=bestfeature.fit(x,y)
```

```python
fit.scores_
```

```python
df=pd.DataFrame(index=x.columns,data=fit.scores_)
```

```python
df
```

```python
df=df.sort_values(by=['scores'],ascending=False).head(15)
```

```python
X=train[df.index]
```

```python
X.columns
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X,y)
```

```python
test=pd.read_csv("test.csv")
```

```python
test=test.apply(le.fit_transform)
```

```python
testx=test[df.index]
```

```python
testx
```

```python
d=reg.predict(testx)
```

```python
output=pd.DataFrame(index=np.arange(1461,2920),data=d)
```

```python
output.columns=['SalePrice']
```

```python
output.to_csv("fsubmit.csv")
```
